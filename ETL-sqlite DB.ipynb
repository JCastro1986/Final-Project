```python
# Import dependencies
import sqlite3
import pandas as pd
```

```python
# get csv to a df
file = pd.read_csv('./Resources/mockData2015.csv')
file.head()
```

```python
# connect to database
conn = sqlite3.connect('./Resources/finalData.db')
c = conn.cursor()
```

```python
# create tables in databse

c.execute('''SELECT count(name) FROM sqlite_master WHERE type='table' AND name='Inmuebles' ''')

#if the count is 1, then table exists
if c.fetchone()[0]==1: 
    print('Table exists.')
else:
    print('Table does not exist.')
    columns = 'id integer, url string, entidad string, calle string, colonia string, \
        municipio string, codigo_postal string, grupo string,tipo string,m2Terreno float,m2Construccion float, \
        recamaras integer,banos integer,estacionamiento integer, pisos integer,cuotaMantto integer,precio float'
    queryCreate = f"CREATE TABLE Inmuebles ({columns})"
    c.execute(queryCreate)
```

```python
df = pd.read_sql_query("SELECT * from Inmuebles", conn)
df.head()
```

```python
# write the data to a sqlite table
file.to_sql("Inmuebles", conn, if_exists='append', index = False)
conn.close()
```
