## Webscrapping from metroscubicos.com

```python
# import instances and libraries
from splinter import Browser
from bs4 import BeautifulSoup as soup
from webdriver_manager.chrome import ChromeDriverManager
import requests
import re
import pandas as pd
#from pathlib import Path
```

```python
# Set up Splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# List all entities available in metroscubicos.com

entidades = ['Aguascalientes', 'Baja California', 'Baja California Sur', 'Campeche', 'Chiapas', 'Chihuahua', 'Coahuila', 'Colima',
            'Distrito Federal', 'Durango', 'Estado De México', 'Guanajuato', 'Guerrero', 'Hidalgo', 'Jalisco', 'Michoacán',
             'Morelos', 'Nayarit', 'Nuevo León', 'Oaxaca', 'Puebla', 'Querétaro', 'Quintana Roo', 'San Luis Potosí',
             'Sinaloa', 'Sonora', 'Tabasco', 'Tamaulipas', 'Tlaxcala', 'Veracruz', 'Yucatán', 'Zacatecas']
```

```python
# make smallcaps and change spaces for dashes
entidadesLower = [x.lower().replace(' ', '-') for x in entidades]
```

```python
entidadesLower = entidadesLower[4:6]
entidadesLower
```

```python
# Initialize variables
cityDict = {}
cityList = []
listLinks = []
baseUrl = 'https://inmuebles.metroscubicos.com/casas/venta/'
```

```python
# Loop para navegar en cada ciudad
for ciudad in entidadesLower:
    links = []
    browseUrl = baseUrl+ciudad
    # visit results for ciudad
    browser.visit(browseUrl)
    html = browser.html
    html_soup = soup(html, 'html.parser')
    # Get number of pages for results
    paginas = html_soup.find('li', class_=re.compile('page-count'))
    numPaginas = [int(s) for s in paginas.get_text().split() if s.isdigit()][0]
    if browser.is_element_present_by_id('newCookieDisclaimerButton'):
        browser.find_by_css('button[id="newCookieDisclaimerButton"]').click()
    # initialize counter for ciudad
    counter = 0
    
    # Browse in each page
    for pagina in range(numPaginas):
        # Get all links
        htmlNew = browser.html
        new_soup = soup(htmlNew, 'html.parser')
        resultados = new_soup.find_all('a', class_="ui-search-link")
        for resultado in resultados:
            if "MLM-" in resultado['href']:
                if resultado['href'] not in links:
                    links.append(resultado['href'])
        # Change page if hasn't reach final
        if (pagina+1) != numPaginas:
            browser.find_by_css('li[class="andes-pagination__button andes-pagination__button--next"] a').first.click()
        
        
    # Loop para links en cada página
    for link in links:
        browser.visit(link)
        asset_soup = soup(browser.html, 'html.parser')
        #initialize internal dict
        dict = {}
        # Save ID
        ID = re.search('(?<=MLM-)\d+',link).group(0)
        # Save link
        urlDict = re.search('[^#]*',link).group(0)
        # Save price
        simbolo = asset_soup.find('span', class_="andes-money-amount__currency-symbol").get_text()
        monto = asset_soup.find('span', class_="andes-money-amount__fraction").get_text()
        precio = simbolo+monto
        # Save location
        divUbicacion = asset_soup.find('div', class_="ui-vip-location")
        ubicacion = divUbicacion.find('p').get_text()
        ubicacion = ubicacion.split(", ")
        # Find table and process columns and rows
        tabla = asset_soup.find('tbody', class_="andes-table__body")
        tablaDatos = asset_soup.find('table', class_="andes-table") 
        nombreDatos = tablaDatos.find_all('th')
        dataDatos = tablaDatos.find_all('td')

        # Save into in dict
        if ID not in dict:
            dict["City"] = ciudad
            dict["ID"] = ID
            dict['Url'] = urlDict
            dict['Price'] = precio
            for idx, dir in enumerate(ubicacion):
                string = "Dir"+ str(idx)
                dict[string] = dir
            for i in range(len(nombreDatos)):
                dict[nombreDatos[i].get_text()] = dataDatos[i].get_text()
        counter += 1
        
        
        # Save dict to list or assets dict
        listLinks.append(dict)

    # Save list of dictionaries to cityDict
    print(f'{ciudad} total {counter} registers')
    #cityDict[ciudad] = listLinks

# Save cityDict to cityList
#cityList.append(cityDict)

browser.quit()
```
